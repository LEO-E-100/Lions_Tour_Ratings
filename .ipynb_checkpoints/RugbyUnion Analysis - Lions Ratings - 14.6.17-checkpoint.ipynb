# r/RugbyUnion Player Ratings Analysis

## Introduction

Data has been taken from the crowd sourced player ratings from each of the games for the British and Irish Lions. Full credit to the subreddit [r/rugbyunion](https://www.reddit.com/r/rugbyunion/) and [u/beneath_the_bridge](https://www.reddit.com/user/beneath_the_bridge/submitted/) in particular for setting up these ratings. Each player in the squad is rated for their performance (assuming they played) after each match.

Currently data is not yet full as not all matches have been played. However the analysis can be put in place now.

The repo for this project can be found [here](https://github.com/LEO-E-100/Lions_Tour_Ratings)

```python
# Import statements and creating the dataframe to be used.

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import linregress
%matplotlib inline

df = pd.read_csv('r_rugbyunion_player_ratings_2.csv')
df.fillna(0, inplace=True)
df['Index'] = df.index
df.head()
```

### Step 1:

First step is to add some feature columns, first off create a "Total Rating" column by summing the value from each game column. In order to do this it is necessary to create a python list of the headings. This list will also be useful later for iterating etc. This can also be used for a sub-dataframe with just the game ratings and no names, clubs etc.

```python
df['Total_Rating'] = (df['NZ_Barbarians'] + df['Blues'] + df['Crusaders'] + df['Highlanders'] + df['Maori_All_Blacks'] + df['Chiefs'] + df['All_Blacks_1'] + df ['Hurricanes'] + df['All_Blacks_2'] + df['All_Blacks_3'])

games = ['NZ_Barbarians', 'Blues', 'Crusaders', 'Highlanders', 'Maori_All_Blacks', 'Chiefs', 'All_Blacks_1', 'Hurricanes', 'All_Blacks_2', 'All_Blacks_3']

df1 = pd.DataFrame(df, columns=games)
df1.head()
```

### Step 2: 

"Total Rating" is useful but a better measure would be an average as this would account for the number of games played. Otherwise a player who only played a few times, for example Stuard Hogg, would be handicapped by only playing a few games.

In order to calculate a mean it is necessary to have a value of the games played. This means testing each column in the "Games" table and if the value is greater than 0 it increments the "Games_Played" value by one. 

This cell throws an annoying exception that is automatically generated, however having looked into it I can't see that it's relevant to the code - I shall ignore for now.

```python
# A total games played column will count the number of non-zero player ratings for each player and store in a new column.
df['Games_Played'] = 0

for i in df['Index']:
    for j in games:
        if df[j][i] != 0.0:
            df['Games_Played'][i] += 1
        else:
            continue
df.head()
```

### Step 3:

Now we can calulate the average rating of players over the course of the tour. This will be placed in the new feature column "Player_Average". 

```python
# Calculating a proper average of player rating per game played
df['Player_Average'] = (df['Total_Rating'])/(df['Games_Played'])
```

### Step 4:

Now we can do some actual data analysis using graphs. First I wanted to see if the number of caps that a player had earned for his home nation made any difference to the average rating per game. 

With just 4 data points so far this was unlikely to be too reliable as a result but as an issue of personal interest this represented a useful starting point.

As the graph shows its a fairly broad scatter that is unlikely to reveal a serious trend.

```python
X = df['National_Caps']
Y = df['Player_Average']

fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot(111)
ax.scatter(X, Y)
ax.set_title("Average Rating by Caps")
ax.set_xlabel ('National Caps')
ax.set_ylabel ('Average Rating')
plt.tight_layout()
plt.show()
fig.savefig("Average_Rating_by_Caps.png")
```

### Step 5:

To test if this graph showed a trend the next step was to create a linear model and run a linRegress. This would show conclusively if any relation existed between the independent and explanatory variable. 

It can be seen both from the graph and the linRegress result that no trend exists. 

#### Linregress Output

The output of a linregress function call is a list of 5 values.

- Slope
- Intercept
- R-value
- P-value
- Standard Error

So it can be seen that the number of caps has very little relevance to the average player rating and no correlation is a fair assumption.

A facile interpretation however suggests that a player with no caps would have an average rating of 6.3 so far this tour.


```python
# A linear model of caps vs. average rating

fit = np.polyfit(X,Y,1)
fit_fn = np.poly1d(fit) 
# fit_fn is now a function which takes in x and returns an estimate for y

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

ax.plot(X,Y, 'yo', X, fit_fn(X), '--k')
ax.set_title("Linear Model of Average Rating by Caps")
ax.set_xlim(0, 120)
ax.set_ylim(3, 10)
ax.set_xlabel ('National Caps')
ax.set_ylabel ('Average Rating')
plt.tight_layout()
plt.show()
fig.savefig("Average_Rating_by_Caps_lm.png")
```

### Step 6:

Finally an actually useful graph. This graph simply produces the average team rating (as calculated by taking an average of each players rating) for each game and then plotting them as a bar graph. 

The most interesting aspect of this is that, from a brief reading of the data it might be expected that a winning game for the Lions would result in higher ratings for the players, while this has happened to an extent it is not nearly as pronounced as might be expected. Obviously this is likely to change as the data gets filled in.

```python
N = 10
ind = np.arange(N)
width = 0.2

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

x = list(df1.columns.values)
np_df1 = df1.as_matrix()
y = np_df1[41]

ax.bar(ind, y, width, color = "#B21020")
plt.title("Average Game Ratings")
ax.set_xlim(-width,len(ind)+width)
ax.set_ylim(0,10)
ax.set_ylabel("Average Team Rating")
ax.set_xlabel("Game")
ax.set_xticks(ind+width)
ax.set_xticklabels(x)
xtickNames = ax.set_xticklabels(x)
plt.setp(xtickNames, rotation=75, fontsize=10)
plt.tight_layout()
plt.show()
fig.savefig("Average_Game_Ratings.png")
```

### Step 7:

As a small step it is useful to create a table of a player and their average rating so far, as this can be useful for a quick glance look at how each player is performing.

```python
small = ["Name", "Player_Average"]
df2 = pd.DataFrame(df, columns=small)
df2.head()
```

### Step 8:

The final graph produced replicates the table above and produces a bar graph of each player's average rating so far. Without published standard error per point these data must be treated with caution, however with just 4 data points the addition of standard error would likely not be clarifying but rather the opposite. 

Overall the key things to look at are the best and worst performers according to these ratings. Conor Murray is a clear outlier of the data, this is likely because he had a stand out game and has only played once. Nowell, Henderson and Williams have clearly struggled this tour and are rated less than 5 on average so far. 

```python
N = 42
ind = np.arange(N)
width = 0.75

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

x = df["Name"]
y = df["Player_Average"]

plt.title("Average Player Ratings")
ax.bar(ind, y, width, color = "#B21020")
ax.set_xlim(-width,len(ind)+width)
ax.set_ylim(0,10)
ax.set_ylabel("Player Rating")
ax.set_xlabel("Player Name")
ax.set_xticks(ind+width)
ax.set_xticklabels(x)
xtickNames = ax.set_xticklabels(x)
plt.setp(xtickNames, rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
fig.savefig("Average_Player_Ratings.png")
```
